# Process-01: <font color='green'>Using pandas to create dummy variables
  This process is easy i like it

```python
dummies = pd.get_dummies(df.town)    # covert town column into dummy variable
dummies
```

```python
# Concat the dummies variable with the main df

merged = pd.concat([df, dummies],axis='columns')
merged
```

**<font color='red'>Dummy Variable Trap**<br>
<font color='green'>1. We need to drop the original town column because we already have these dummy variable columns. we don't need this text column anyway because that's not go to work on our linear regressing model.

<font color='green'>2. We have to drop one of these dummy variables column. Now why do we want to do it? You can Google about 'dummy variable trap' and what that concept means it whenever one variable can be derived from a rest of the variables these variable are said to be multi-colinear and whenever you have multi-colinearity in your dataset it create the problem of dummy variable trap that mess up machine learing model.**

<font color='green'>Hence the rule of the thumb base you have to drop one the dummy variable colunmsso if you have five dummy variables you can drop any one.

<font color='green'>Here we three dummy variables so we have to drop any of them</font>
    
 ------------------------------------------------------------------------------------------------------------------------------
 ------------------------------------------------------------------------------------------------------------------------------
    
When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need to drop one column.

**NOTE:**<br> sklearn library takes care of dummy variable trap hence even if you don't drop one of the state columns it is going to work, however we should make a habit of taking care of dummy variable trap ourselves just in case library that you are using is not handling this for you

```python
final_df = merged.drop(['town', 'west windsor'], axis='columns')
final_df
```

```python
from sklearn import linear_model as lm
model = lm.LinearRegression()
```

**As price is an independent variable , So it no need to keep in the df. And hence we have to drop the price column from the df and keep it in an another independent variable** 


```python
X = final_df.drop('price',axis='columns')
X
```

```python
y = final_df.price
y
```

```python
model.fit(X,y)
```

**1) 2800 sqr ft home in robinsville area**

```python
model.predict([[2800, 0, 1]])
```

**2) 3400 sqr ft home in monroe township area**

```python
model.predict([[3400, 0, 0]])
```

## Accuracy calculation

```python
model.score(X,y)   # Our model is 95% accurate
```

# Process-02: <font color='green'>Using OneHotEncoding of sklearn
    
First step is to use label encoder to convert town names into numbers

```python
from sklearn import preprocessing as pp #from sklearn.preprocessing import LabelEncoder
le = pp.LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['town', 'area']].values        # .value use korle df return na kore 2d array return kore
X
```

```python
y = dfle.price.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(),[0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:, 1:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[1,0,2800]])
```

```python
model.predict([[0,1,3400]])
```

```python
import pandas as pd
df_hw = pd.read_csv('HW_carprices.csv')
df_hw
```

### Using pandas to create dummy variables

```python
dummies  = pd.get_dummies(df_hw['Car Model'])
```

```python
dummies
```

```python
marged = pd.concat([df_hw, dummies], axis = 'columns')
marged
```

```python
marged = marged.drop(['Car Model'], axis='columns')
marged
```

**Dummy Variable Trap**<br>
When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need to drop one column.

**NOTE:**<br> sklearn library takes care of dummy variable trap hence even if you don't drop one of the state columns it is going to work, however we should make a habit of taking care of dummy variable trap ourselves just in case library that you are using is not handling this for you

```python
final_df = marged.drop(['Mercedez Benz C class'], axis = 'columns') 
final_df                                                  # marged.drop(['Car Model','Mercedez Benz C class'], axis = 'columns')
```

**As Sell Price is an dependent variable , So it no need to keep in the df.  And hence we have to drop the price column from the df and keep it in an another dependent variable**

```python
X = final_df.drop(['Sell Price($)'], axis = 'columns')
X
```

```python
y = final_df['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)   # Here X is an independent variable and y is a target variable
```

```python
model.predict(X)
```

**1) Price of mercedez benz that is 4 yr old with mileage 45000**

```python
model.predict([[45000, 4, 0, 0]])
```

**2) Price of BMW X5 that is 7 yr old with mileage 86000**

```python
model.predict([[86000, 7, 0, 1]])
```

**3) Tell me the score (accuracy) of your model.**
