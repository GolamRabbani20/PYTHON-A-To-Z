# <font color="red">Introduction ot Matplotlib</font>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5, 6, 7]            # week days
y = [50, 55, 45, 90, 80, 25, 20]     #Temperature
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, color="blue")
```

```python
plt.xlabel("Day")
plt.ylabel("Temperatures")
plt.title('Weater details of a week')
plt.plot(x, y, color="green")
```

```python
plt.xlabel("Day")
plt.ylabel("Temperatures")
plt.title('Weater details of a week')
plt.plot(x, y, color="green", linewidth=5)
```

```python
plt.xlabel('Days')
plt.ylabel('Temperatures')
plt.title('Weather details')
plt.plot(x, y, color='#ff0509', linewidth=3, linestyle='dotted')
```
