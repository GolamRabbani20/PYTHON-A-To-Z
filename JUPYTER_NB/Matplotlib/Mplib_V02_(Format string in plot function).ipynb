# <font color="red">Format strings in plot function</font>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [78, 95, 70, 80, 58, 66, 98]
plt.plot(x, y)
```

```python
plt.plot(x, y, '--')
```

```python
plt.plot(x, y, '--r')  # r= red color
```

```python
plt.plot(x, y, '+--')
```

```python
plt.plot(x, y, 'D--') #D = diamond
```

```python
plt.plot(x, y, 'gD--') # green diamond
```

```python
plt.plot(x, y, '+')
```

```python
plt.plot(x, y, 'D')
```

```python
plt.plot(x, y, 'rD')
```

```python
plt.plot(x, y, color='red', marker='D', linestyle='')
```

```python
plt.plot(x, y, color='c', marker='o', markersize=10, linestyle='dotted')
```

```python
plt.plot(x, y, color='red', marker='D', alpha=.4)
```
