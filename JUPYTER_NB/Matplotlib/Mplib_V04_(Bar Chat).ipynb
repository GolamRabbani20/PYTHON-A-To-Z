# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['Google', 'Amazone', 'MSFT', 'FB']
revenue = [45, 89, 25, 70]
profit = [24, 30, 19, 40]
```

```python
plt.xlabel("Company")
plt.ylabel("Revenue(bln)")
plt.title("Company vs Revenue")
plt.bar(company, revenue)
```

```python
xpos =  np.arange(len(company)) # ypos = Y axis position
plt.bar(ypos, revenue)
```

```python
plt. xticks(ypos, company)
plt.xlabel("Company")
plt.ylabel("Revenue(bln)")
plt.bar(ypos, revenue, label="Revenue")
plt.legend()
```

```python
plt.ylabel("revenue")
plt.xlabel('company')
plt.title('US tech stock')
plt.bar(xpos-0.2, revenue,  width=0.4, label="revenue")
plt.bar(xpos+0.2, profit, width=0.4,label="profit")
plt.legend()
```

```python
plt. yticks(ypos, company)
plt.xlabel("profit")
plt.ylabel("company")
plt.title('US tech stock')
plt.barh(xpos-0.2, revenue,height=0.4, label="revenue")
plt.barh(xpos+0.2, profit,height=0.4, label="profit")
plt.legend()
```
