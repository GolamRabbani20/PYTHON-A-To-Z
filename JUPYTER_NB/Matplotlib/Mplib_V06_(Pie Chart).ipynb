# <font color="red">Pie Chart</font>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_values = [1400, 600, 300, 410, 250]
exp_labels = ['Home rent', 'Food', 'Internet bill', 'car','Other Utilities']
```

```python
plt.pie(exp_values, labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_values, labels=exp_labels, radius=2, autopct='%0.2f%%', shadow=True, explode=[0,0.2,0,0.3,0], startangle=45)
plt.show
```
