# <font color='red'>Save Chart</font>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_values= [1400, 600, 300, 410, 250]
exp_labels = ['Home rent', 'Food', 'Internet bill', 'car','Other Utilities']
```

```python
plt.axis("equal")
plt.pie(exp_values, labels=exp_labels, radius=3, autopct='%0.2f%%', shadow=True, explode=[0,0,0,0.1,0], startangle=45)
plt.savefig("D:\Matplotlib_figure\piechart.png", bbox_inches='tight', pad_inches=2, transparent=True)
```
