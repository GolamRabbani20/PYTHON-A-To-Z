# Concat DataFrame

```python
import pandas as pd
bd_weather = pd.DataFrame({
    'city':['Dhaka', 'Rongpur', 'Nilphamri', 'Jossor', 'Khulna'],
     'temperature':[45,50,25,60,40],
     'humidity':[60,70,56,40,78]
}) 
bd_weather
```

```python
india_weather = pd.DataFrame({
    'city':['Dhilli', 'Kalkata', 'Mubbai', 'Kasmir', 'Noyha-dhilli'],
    'temperature':[50, 60, 47, 75, 58],
    'humidity':[10, 15, 85, 69, 32]
})
india_weather
```

```python
usa_weather = pd.DataFrame({
    'city':['New-York', 'Washington', 'Chicago', 'Boston', 'Austin'],
    'temperature':[50, 79, 80, 90, 65],
    'humidity':[54, 58, 69, 13, 25]
})
usa_weather
```

```python
df = pd.concat([bd_weather, india_weather, usa_weather])
df
```

```python
new_df = pd.concat([bd_weather, india_weather, usa_weather], ignore_index=True)
new_df
```

```python
df = pd.concat([bd_weather, india_weather, usa_weather], keys=["Bangladesh", "India", "USA"])
df
```

```python
df.loc['Bangladesh']
```

```python
df.loc["USA"]
```

```python
teperature_bd = pd.DataFrame({
    'city':['Dhaka', 'Rongpur', 'Nilphamri', 'Jossor', 'Khulna'],
    'temperature':[45,50,25,60,40]}, index=[0,1,2,3,4])
teperature_bd
```

```python
windspeed_bd = pd.DataFrame({
    'city':['Rongpur', 'Khulna', 'Nilphamri','Dhaka', 'Jossor'],
    'windspeed':[4,7,20,6,12]},index=[1,4,2,0,3])
windspeed_bd
```

```python
df = pd.concat([teperature_bd, windspeed_bd])
df
```

```python
df = pd.concat([teperature_bd, windspeed_bd], axis=1)
df
```

```python
s= pd.Series(['Rain','Sunny', 'Snow', 'Rain','Sunny'], name='Event')
s
```

```python
df= pd.concat([df, s], axis=1)
df
```
