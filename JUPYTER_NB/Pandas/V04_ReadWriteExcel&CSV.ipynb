```python
import pandas as pd
def Change_people(cell):
    if cell=='bill gates':
        return 'Rabbani'
    return cell
df=pd.read_excel("F:\JUPYTER_NB\A_Excel&CSV\V4_stock_data.xlsx","Sheet1", converters={'people':Change_people})
df
```

```python
def Change_People_Cell(x):
    if x == "Rabbani":
        return "bill gates"
    return x
def change_eps_data(cell):
    if cell=="not available":
        return None
    return cell
df = pd.read_excel("F:\JUPYTER_NB\A_Excel&CSV\V4_stock_data.xlsx","Sheet1" , converters={
    'people':Change_People_Cell,
     'eps': change_eps_data
})
df
```

```python
weather_data=pd.DataFrame({  
    'day':['20/01/21', '21/01/21', '22/01/21', '23/01/21', '24/01/21', '25/01/21', '26/01/21', '27/01/21'],
    'Temperature': [32, 35, 25, 26, 45, 20, 12, 22],
    'Windspeed':[6,7,8,10,9,5,4,6],
    'Event': ['Rain', 'Sunny', 'Rain', 'Snow', 'Snow', 'Rain', 'Sunny', 'Snow']
})

Stock_data=pd.DataFrame({
                        'Tricket':['GOOGLE', 'WMT', 'MSFT', 'RIL', "TATA"],
                        'EPS':[27.82, 4.61, -1,'NaN', 5.6],
                        "Revenue":[87, 484, 85, 50, -1],
                        "Price":[845, 65, 64, 1023, "n.a."],
                        'Peopel':['Larry Page', 'n.a.', 'Bill Gates', 'Mukesh Ambani', 'Ratan Tata']
                      })

with pd.ExcelWriter('Stocks_&_Weather.xlsx') as writer:
    weather_data.to_excel(writer, sheet_name="Weather", index=False)
    Stock_data.to_excel(writer, sheet_name="Stocks")
```
