# <font color='blue'>Use of replace function in pandas and numpy</font>

```python
import pandas as pd
import numpy as np
df= pd.read_csv('weather_data.csv')
df
```

```python
new_df=df.replace([-99999,-8888], np.NaN)
new_df
```

```python
new_df=df.replace({
    'temperature':[-99999, -8888],
    'windspeed':[-99999, -8888],
    'event':'0'
},np.NaN)
new_df
```

```python
new_df=df.replace({
    -99999:np.NaN,
    '0':'Sunny'
})
new_df
```

```python
import pandas as pd
import numpy as np
df= pd.read_csv('V6_weather_data.csv')
df
```

```python
new_df=df.replace('[A-Za-z]','',regex=True)
new_df
```

```python
new_df=df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
new_df
```
