# Merge Dataframe in pandas

```python
import pandas as pd
df1 = pd.DataFrame({
    'city': ['Dhaka', 'Rongpur', 'Khulna', 'Saidpur', 'Rajshahi'],
    'Temperature': [12, 25, 45, 36, 50]
})
df1
```

```python
df2 = pd.DataFrame({
    'city': ['Khulna', 'Rajshahi', 'Dhaka', 'Rongpur', 'Saidpur'],
     'Humidity': [12, 25, 50, 45, 70]
})
df2
```

***<font color="red">Inner join ---> Inter section of two or more tables***</font>

```python
df3 = pd.merge(df1, df2, on='city') #By defualt how= inner
df3
```

***<font color="red">Outer join ----> Unioun of two or more tables***</font>

```python
df4 = pd.DataFrame({
    'city': ['Dhaka', 'Rongpur', 'Khulna', 'Saidpur', 'Savar'],
    'Temperature': [12, 25, 45, 36, 50]
})
df4
```

```python
df5 = pd.DataFrame({
    'city': ['Khulna', 'Rajshahi', 'Dhaka', 'Borishal', 'Saidpur'],
     'Humidity': [12, 25, 50, 45, 70]
})
df5
```

```python
df6 = pd.merge(df4, df5, how='outer') #By defualt how=inner, how=[inner, outer, left, right]
df6
```

```python
df60 = pd.merge(df4, df5, how='outer', indicator=True) # Indicator flag
df60
```

***Common columns in dataframe***

```python
df7 = pd.DataFrame({
    'city': ['Dhaka', 'Rongpur', 'Nilphamari'],
    'Temperature': [40, 100, 49],
    'Humidity': [45, 58, 69]
})
df7
```

```python
df8 = pd.DataFrame({
    'city': ['Nilphamari', 'Dhaka', 'Rongpur'],
    'Humidity': [15, 58, 69],
    'Temperature': [90, 100, 80]
})
df8
```

```python
df9 = pd.merge(df7, df8 , on='city')
df9
```

```python
df10 = pd.merge(df7, df8, on='city', suffixes=('_left-DF', '_right-DF'))
df10
```
