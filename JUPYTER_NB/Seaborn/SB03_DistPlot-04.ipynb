```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
penguins = sns.load_dataset('penguins')
penguins
```

```python
penguins.head()
```

```python
 sns.distplot(penguins.flipper_length_mm)
```

```python
sns.distplot(penguins.flipper_length_mm, rug = True)
```

```python
 sns.distplot(penguins.flipper_length_mm, hist = False)
```

#### Setting the number of bins

```python
sns.distplot(penguins.flipper_length_mm, bins = 20)
```

#### Setting the bin values as an array

```python
sns.distplot(penguins.flipper_length_mm, bins = [160, 170, 180, 190, 200, 240])
```

#### Fitting the data to a paricular distribution using norm_hist

```python
 sns.distplot(penguins.flipper_length_mm, norm_hist = True, kde = False)
```

#### Styling Rugplot

```python
sns.distplot(penguins.flipper_length_mm, rug_kws = {'color': 'red','height': 0.30}, rug = True)
```

#### Styling Histograms

```python
sns.distplot(penguins.flipper_length_mm, kde = False, hist_kws = {'color':'blue', 'alpha': 0.3, 'histtype': 'step'})
```

```python
sns.distplot(penguins.flipper_length_mm, kde = False, hist_kws = {'color':'blue', 'alpha': 0.2, 'histtype': 'stepfilled'})
```

#### Styling KDE Plots

```python
sns.distplot(penguins.flipper_length_mm, kde = True, hist = False, kde_kws = {'color':'green', 'lw': 2.5, 'alpha': 0.4})
```
